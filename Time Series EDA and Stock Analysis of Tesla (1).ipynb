## Time Resampling

- Time resampling refers to the process of changing the frequency of your time series data, such as converting daily data into monthly data or vice versa, to facilitate analysis and comparison at different time intervals.

#### Rule A

- *Annual resampling:* Aggregates data at the end of each calendar year.
- Example: Summarizing Tesla's annual stock performance.

#### Rule QS

- *Quarter Start resampling:* Aggregates data at the start of each calendar quarter.
- Example: Evaluating Tesla's stock data from the beginning of each quarter.
  
#### Rule BA

- *Business Year End resampling:* Aggregates data at the end of each business (fiscal) year.
- Example: Reviewing Tesla's fiscal year-end financial metrics.
  
#### Rule BQS

- *Business Quarter Start resampling:* Aggregates data at the start of each business quarter.
- Example: Analyzing Tesla's stock performance at the beginning of each fiscal quarter.

```python
#minimum/lowest stock values
tsla_data.resample(rule='A').min()
```

```python
#maximum/Higest stock values
tsla_data.resample(rule='A').max()
```

```python
#Rule A - means year & frequency
# Raise in stock price from 2010 to 2024
tsla_data.resample(rule='A').max()['Open'].plot()
```

```python
# https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78 (Document for reference)
# rule QS - quarterly start frequency
tsla_data.resample(rule='QS').max()['High'].plot()
```

```python
# Rule BA - Business end frequency
tsla_data.resample(rule='BA').max()
```

```python
# Rule BQS - Business Quarters
tsla_data.resample(rule='BQS').max()
```

```python
# plotting bar charts
tsla_data['Open'].resample(rule='A').mean().plot(kind='bar')
```

```python
# plotting bar charts & figure size
tsla_data['Open'].resample(rule='BQS').max().plot(kind='bar',figsize=(20,5))
```

### Rolling

- Rolling refers to applying a sliding window function over a series of data points. Common aggregate functions used with rolling include:

- Mean: Computes the average value over the window.
- Sum: Calculates the total sum of values in the window.
- Min: Finds the minimum value in the window.
- Max: Finds the maximum value in the window.
- Std: Computes the standard deviation of values in the window.

- Rolling is a technique of applying a function (such as mean, sum, min, max, etc.) to a sliding window of data points in a time series. This sliding window moves over the data one observation at a time, calculating the desired function for each window. This allows analysts to compute metrics over a specific period or window of time, providing insights into trends and patterns that might not be evident when looking at the data in its entirety.

*For example*, if you have daily stock prices over several years, you can compute a 30-day rolling average by taking the average of the stock prices for each consecutive 30-day period. This rolling average helps smooth out short-term fluctuations and highlights longer-term trends in the stock's price movements.

- Rolling calculations are commonly used in time series analysis for tasks such as trend identification, volatility estimation, and anomaly detection. They provide a flexible way to analyze data over different time periods without losing the sequential structure inherent in time series data.


```python
# rolling
tsla_data['High'].rolling(10)
```

```python
#calculating the rolling mean
tsla_data['High'].rolling(11).mean().head(21)
```

```python
#calculating the rolling max
tsla_data['High'].rolling(15).max().head(21)
```

```python
# rolling mean average of stock prices over a 30-day window
tsla_data['Open:30 days rolling']=tsla_data['Open'].rolling(30).mean()
```

#### Let's break down what each part of the code does: 

- tsla_data['Open:30 days rolling']: This specifies the column name where the results of the rolling mean calculation will be stored. In this case, it's creating a new column named 'Open:30 days rolling' in the tsla_data DataFrame.

- = tsla_data['Open'].rolling(30).mean(): tsla_data['Open']: This selects the 'Open' column from the tsla_data DataFrame, which contains the daily opening prices of Tesla stock.

- .rolling(30): This is the rolling window function provided by pandas. It specifies that we want to calculate statistics (in this case, the mean) over a rolling window of 30 periods.
  
- .mean(): This specifies the aggregation function to apply to each rolling window. Here, it computes the mean (average) of the values in each 30-day window.
  
#### Explanation:

- The code calculates a new column 'Open:30 days rolling' where each value is the average of the 'Open' prices over the previous 30 days.
- This rolling mean calculation helps to smooth out short-term fluctuations in the daily opening prices and provides a clearer picture of the underlying trend in the data over time.
- It's important to note that the first 29 values in the 'Open:30 days rolling' column will be NaN (Not a Number) because there isn't enough data to compute a full 30-day rolling mean until the 30th day.
- This technique is useful for time series analysis to identify trends, seasonal patterns, or anomalies in the data.
- In summary, the code segment tsla_data['Open:30 days rolling'] = tsla_data['Open'].rolling(30).mean() computes a 30-day rolling average of the 'Open' prices for Tesla stock and stores the results in a new column 'Open:30 days rolling' in the tsla_data DataFrame.

- In rolling calculations in data analysis, *"NaN"* stands for *"Not a Number."* It represents missing or undefined values in the data, which can occur when there is no data available or when operations result in undefined numerical outcomes, such as division by zero. When performing rolling calculations, handling NaN values appropriately is crucial for accurate analysis and visualization of data trends over time.
